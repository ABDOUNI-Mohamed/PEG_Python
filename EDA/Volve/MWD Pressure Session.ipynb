#### Let's collect all the LAS files

```python
files = []
path = r"M:\Volve\Well_logs\03.PRESSURE\15_9-F-1\\"
preffix = "FM_PRESS_RAW_RUN4_MWD_"
suffix = ".las"
n_files = 6
for i in range(1,n_files+1):
    files.append(path+preffix+str(i)+suffix)
```

```python
files
```

```python
color = plt.cm.rainbow([0,1,2,3,4,5])
color
```

```python

for i, file in enumerate(files):
    fig, axes = plt.subplots(1, 3, figsize=(20,5))
    las1 = lasio.read(file)
    axes[0].plot(las1['TIME'], las1['DD_VOLUME'], color=color[i])
    axes[0].set_title('Volume for Station ' + str(i+1))
    axes[1].plot(las1['TIME'], las1['DD_PRESS'], color=color[i])
    axes[2].plot(las1['TIME'], las1['DD_PRESS_MOD_1'], color=color[i])
    plt.show()
```
