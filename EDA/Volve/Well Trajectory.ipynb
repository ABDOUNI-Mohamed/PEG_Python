```python
def casting_func(x):
    if isinstance(x, float):
        pass
    elif "'" in x:
        return x
    elif '"' in x:
        return np.nan 
    elif ":" in x:
        print('here we go')
        return np.nan       
    else:
        x = x.replace(' ','')
        x = x.replace('.','')
        x = x.replace(',','.')
        return float(x)
    
def well_deviation_plot(dev, datum):
    np.set_printoptions(suppress=True, precision=2)
    z = welly.Well()
    z.location = welly.Location(params={'kb': 0}) #params={'kb': 10})
    z.location.add_deviation(dev) #, td=1000, azimuth_datum=20)
    #z.location.trajectory(datum=datum, elev=True)
    #fig, ax = plt.subplots(figsize=(12, 7), subplot_kw={'projection': '3d'})
    #ax.plot(*z.location.trajectory().T, lw=3, alpha=0.75)
    return z#grf# z.location.plot_3d()
```

```python
tablex = []
deviation_surveys = []
for i, well in enumerate(wells_info):
    well_survey = {}
    df = pd.DataFrame()
    file = wells_info[well]['path']
    for j, page in enumerate(wells_info[well]['pages']):
        tables = tabula.read_pdf(file, pages = str(page), multiple_tables = True)
        tablex = tables[-1]
        tablex = pd.DataFrame(np.vstack([tablex.columns, tablex]))
        df = pd.concat([df, tablex], axis=0, ignore_index=True)
    df = df.applymap(casting_func)
    df = df.dropna()
    datum = df.loc[0,[6,7,3]].to_numpy()
    datum = [int(datum[0]), int(datum[1]), datum[2]]
    z = well_deviation_plot(df.loc[:,[0,1,2]].to_numpy(), datum)
    well_survey['well_name'] = well
    well_survey['deviation'] = z
    well_survey['datum'] = datum
    deviation_surveys.append(well_survey)  
```

```python
#100 colors
color=cm.rainbow(np.linspace(0,1,100))
```

```python
fig, ax = plt.subplots(figsize=(15, 10), subplot_kw={'projection': '3d'})

for i, well_survey in enumerate(deviation_surveys):
    ax.plot(*well_survey['deviation'].location.trajectory(datum=well_survey['datum'], elev=True).T, lw=3, alpha=0.75, color=color[i], label=well_survey['well_name'])
plt.legend()
plt.show()
```
