```python
#100 colors
color=cm.rainbow(np.linspace(0,1,100))

files = []
path = r"M:\Volve\Well_logs\03.PRESSURE\15_9-F-1\\"
root_name = "FM_PRESS_RAW_RUN4_MWD_"
extension = ".LAS"
number_of_files = 6
for i in range(1, number_of_files+1):
    file = root_name + str(i) + extension
    files.append(file)
```

```python
color[1]
```

```python
for i, file in enumerate(files):
    print('Station ', i+1)
    las1 = lasio.read(path+file)
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))
    axes[0].plot(las1['TIME'], las1['DD_PRESS'], label=file, color=color[i])
    axes[1].plot(las1['TIME'], las1['DD_VOLUME'], label=file, color=color[i])
    axes[2].plot(las1['TIME'], las1['DD_PRESS_MOD_1'], label=file, color=color[i])
    plt.title(file)
    plt.show()
```
