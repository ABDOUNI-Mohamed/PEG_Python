```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
indeces = np.arange(0,73,1)
cases_path = r'./models'
prefix = 'CASE_REV3_EX3_'
extension = '_kpi.csv'
cases_names = []
for index in indeces:
    cases_names.append(cases_path+'/'+prefix+str(index)+extension)
#cases_names
df = pd.DataFrame()
existing_indeces = []
for i, case in enumerate(cases_names):
    if os.path.isfile(case):
        existing_indeces.append(i)
        df = pd.concat([df, pd.read_csv(case)])
    else:
        pass
df['existing_indeces'] = existing_indeces
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.set_index(pd.Series(existing_indeces), inplace=True);
df
df['index_corrected'] = df.index
```

```python
df
```

```python
path = "DOE_EX3_with_index.csv"
df2 = pd.read_csv(path)
df2
```

```python
df2['index_corrected'] = [i-1 for i in df2['index']]
mob_ratio_corrected = []
for x in df2['mob_ratio']:
    if x == 'low':
        mob_ratio_corrected.append(1)
    elif x == 'mid':
        mob_ratio_corrected.append(2)
    elif x == 'high':
        mob_ratio_corrected.append(3)
df2['mob_ratio_corrected'] = mob_ratio_corrected
```

```python
df2
```

```python
df3 = df2.merge(df, on='index_corrected')
```

```python
df3
```

### Select Features, Split data into train and test, fit model

```python
df3 = df3[df3['heterogeneity_index']!='zero']
df3['heterogeneity_index'] = [int(i) for i in df3['heterogeneity_index']]
columns_to_keep = ['heterogeneity_index', 'variogram_angle', 'max_lateral_connectivity', 'mob_ratio_corrected', 'mean_permeability', 'kvhk']
y = df3['EUR']
df3 = df3[columns_to_keep]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df3, y, test_size=0.2, random_state=1)
X_test
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_predict = lm.predict(X_test)
```

```python
plt.scatter(y_predict, y_test)
```
